1. Інкапсуляція

```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  # публічний атрибут
        self.__balance = balance  # приватний атрибут

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposit: +{amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdraw: -{amount}")
        else:
            print(f"Withdraw: -{amount} (Insufficient funds)")

    def get_balance(self):
        return self.__balance


# Створення рахунку
account = BankAccount("Bohdan", 1000)

# Виклик методів у циклі з випадковими числами
for _ in range(10):
    amount = random.randint(50, 500)
    
    if random.choice([True, False]):
        account.deposit(amount)
    else:
        account.withdraw(amount)

# Кінцевий баланс
print(f"\nFinal balance: {account.get_balance()}")
```

2. Наслідування

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def start_engine(self):  # новий метод
        return f"{self.brand} {self.model} engine started 🚗"


class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)  # виклик конструктора базового класу
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"


car = Car("Toyota", "Camry", 5)

print(car.display_info())      # перевизначений метод
print(car.start_engine())      # виклик методу з базового класу
```

3. Поліморфізм

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Fish(Animal):
    pass  # метод speak не перевизначений


animals = [Dog(), Cat(), Fish()]

for animal in animals:
    print(animal.speak())
```

4. Абстракція

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area())
```
