Інкапсуляція

```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  
        self.__balance = balance 

    def deposit(self, amount):
        """Поповнення рахунку"""
        self.__balance += amount
        return f"Поповнено: +{amount} грн"

    def withdraw(self, amount):
        """Зняття коштів"""
        if amount <= self.__balance:
            self.__balance -= amount
            return f"Знято: -{amount} грн"
        else:
            return f"Недостатньо коштів! Спроба зняти {amount} грн"

    def get_balance(self):
        """Отримання поточного балансу"""
        return self.__balance
    
    def show_info(self):
        """Виведення інформації про рахунок"""
        return f"Власник: {self.owner}, Баланс: {self.__balance} грн"


print("="*70)
print("СИМУЛЯЦІЯ БАНКІВСЬКОГО РАХУНКУ")
print("="*70)

account = BankAccount("Bohdan", 1000)
print(f"\nПочатковий стан:")
print(account.show_info())

print("\n" + "="*70)
print("ВИПАДКОВІ ОПЕРАЦІЇ (10 ітерацій)")
print("="*70)

num_operations = 10

for i in range(1, num_operations + 1):

    operation = random.randint(0, 1)
    
    amount = random.randint(50, 500)
    
    print(f"\n[Операція {i}]")
    
    if operation == 0:
        result = account.deposit(amount)
        print(f"  Дія: Поповнення рахунку")
        print(f"  {result}")
    else:
        result = account.withdraw(amount)
        print(f"  Дія: Зняття з рахунку")
        print(f"  {result}")
    
    print(f"  Поточний баланс: {account.get_balance()} грн")

print("\n" + "="*70)
print("КІНЦЕВИЙ РЕЗУЛЬТАТ")
print("="*70)
print(f"\nПочатковий баланс: 1000 грн")
print(f"Кінцевий баланс:   {account.get_balance()} грн")
print(f"Зміна:             {account.get_balance() - 1000:+d} грн")
print(f"\n{account.show_info()}")
print("\n" + "="*70)
```

Використання парадигм для створення простої гри

```python
from abc import ABC, abstractmethod
from random import randint, choice


class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health
    
    @abstractmethod
    def attack(self, another_item):
        pass
    
    @abstractmethod
    def boost(self):
        pass
    
    def is_alive(self):
        return self.health > 0


class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"Завдаємо удару мечем {self.name} та наносимо {current_attack} шкоди. У {another_item.name} залишилось здоров'я: {another_item.health}"
    
    def boost(self):
        self._sharp += 1
        return f"{self.name} заточено. Гострота +{self._sharp}"
    
    def sharpening(self):
        self._sharp += 1
    
    @property
    def get_attack_power(self):
        return f"Атака меча {self.name}: {self.__attack_power + self._sharp} одиниць"


class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(0, 20)
        another_item.health -= current_attack
        return f"Завдаємо удару сокирою {self.name} та наносимо {current_attack} шкоди. У {another_item.name} залишилось здоров'я: {another_item.health}"
    
    def boost(self):
        self._sharp += 2
        return f"{self.name} входить у лють. Сила +{self._sharp}"
    
    @property
    def get_attack_power(self):
        return f"Атака сокири {self.name}: {self.__attack_power + self._sharp} одиниць"


class Bow(Item):
    def __init__(self, name, attack_power: int, range_power: int = 5):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._range_power = range_power
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(5, 15) + self._range_power
        another_item.health -= current_attack
        return f"Завдаємо постріл з лука {self.name} та наносимо {current_attack} шкоди. У {another_item.name} залишилось здоров'я: {another_item.health}"
    
    def boost(self):
        self._range_power += 1
        return f"{self.name} перезаряджено. Дальність +{self._range_power}"
    
    def reload(self):
        self._range_power += 1
        return f"{self.name} перезаряджено. Дальність: {self._range_power}"
    
    @property
    def get_attack_power(self):
        return f"Атака лука {self.name}: {self.__attack_power + self._range_power} одиниць"


def create_random_weapon():
    weapons = [
        Sword("Екскалібур", 100),
        Axe("Кратос", 95),
        Bow("Соколине Око", 85, 10)
    ]
    return choice(weapons)


S = Sword("Екскалібур", 100)
A = Axe("Кратос", 95)
B = Bow("Соколине Око", 85, 10)

all_weapons = [S, A, B]
player_weapon = choice(all_weapons)
enemy_weapon = choice(all_weapons)

print("Розпочинаємо гру!")
print(f"Ваша зброя: {player_weapon.name}")
print(f"Зброя противника: {enemy_weapon.name}")
print()

for i in range(20):
    print(f"Хід {i+1}")
    print("-" * 70)
    
    player_action = randint(1, 10)
    if player_action <= 7:
        print(player_weapon.attack(enemy_weapon))
    else:
        print(player_weapon.boost())
        print("Ви підсилили зброю замість атаки")
    
    if not enemy_weapon.is_alive():
        print(f"\nПеремога за {player_weapon.name}")
        break
    
    enemy_action = randint(1, 10)
    if enemy_action <= 7:
        print(enemy_weapon.attack(player_weapon))
    else:
        print(enemy_weapon.boost())
        print("Противник підсилив зброю замість атаки")
    
    if not player_weapon.is_alive():
        print(f"\nПеремога за {enemy_weapon.name}")
        break
    
    print()

print()
print("=" * 70)
print("Кінцева статистика:")
print(f"{player_weapon.name}: {player_weapon.health} HP | {player_weapon.get_attack_power}")
print(f"{enemy_weapon.name}: {enemy_weapon.health} HP | {enemy_weapon.get_attack_power}")
print("=" * 70)
```

Робота з properties

```python
from datetime import datetime

class MyAnimals:
    max_age = 20
    
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    @property
    def age_category(self):
        if self.age < 2:
            return "Малюк"
        elif self.age < 7:
            return "Дорослий"
        elif self.age < 12:
            return "Літній"
        else:
            return "Дуже літній"
    
    @property
    def info(self):
        return f"{self.animals} {self.name}, {self.age} років ({self.age_category})"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"


a1 = MyAnimals("Кіт", "Мурка", 5)
a2 = MyAnimals("Собака", "Рекс", 1)
a3 = MyAnimals("Кіт", "Барсик", 10)

print("Виведення properties для a1:")
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Категорія віку: {a1.age_category}")
print(f"Повна інформація: {a1.info}")

print("\nДодаємо один рік:")
a1.add_one_year()
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Категорія віку: {a1.age_category}")
print(f"Повна інформація: {a1.info}")

print("\n" + "="*70)
print("Виведення properties для всіх об'єктів:")
print("="*70)

for animal in [a1, a2, a3]:
    print(f"\n{animal.info}")
    print(f"  Буде жити до: {animal.live_untill} року")
    print(f"  Категорія: {animal.age_category}")
```

Методи Класу

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1
        self.validate_name()

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)
    
    @classmethod
    def validate_name_format(cls, name):
        if not name[0].isupper():
            return False
        return True
    
    def validate_name(self):
        if not self.name[0].isupper():
            print(f"Попередження: Ім'я '{self.name}' не починається з великої букви!")
            self.name = self.name.capitalize()
            print(f"Виправлено на: '{self.name}'")
        else:
            print(f"Ім'я '{self.name}' задано коректно (з великої букви)")
        return self.name


a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)
a3 = MyAnimals("Кіт", "барсик", 3)
a4 = MyAnimals("Папуга", "КЕША", 1)

print("\n" + "="*70)
print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Мій {a3.animals}, звати {a3.name}, має {a3.age} років")
print(f"Мій {a4.animals}, звати {a4.name}, має {a4.age} років")
print("="*70)

print(f"\nЗагальна кількість об'єктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")

print("\n" + "="*70)
print("Перевірка імен через класовий метод:")
print("="*70)
test_names = ["Мурка", "рекс", "БАРСИК", "кеша"]
for test_name in test_names:
    result = MyAnimals.validate_name_format(test_name)
    status = "Коректно" if result else "Некоректно"
    print(f"Ім'я '{test_name}': {status}")
```

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```

```python
a = 2
b = 5
c = "C"

print(a+b, a - int(1),  a * b, str(a) + c)

print(a.__add__(b), a.__sub__(int(1)), a.__mul__(b))

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"Мій {self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            self.age += obj
            return f"Мій {self.name} постарів на {obj} років, тепер йому {self.age}"
        return NotImplementedError
    
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"Старшим є {self.name}"
            return f"{self.name} не є старшим за {obj.name}"
        return NotImplementedError

    def __len__(self) -> int:
        print("Ми зайшли в метон len, та знаходимо довжину Імені тваринки")
        return len(self.name)
            
        

a = MyAnimals("Собака", "Рекс", 2)
b = MyAnimals("Кіт", "Мурзик", 4)
print(a - b)
print(a - 1)
print(a - float(1))

print(a > b)

print(b > a)

len(a)
```
