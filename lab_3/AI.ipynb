Приклад відповіді від AI:
 1. Типи даних і змінні

```python
a = "Привіт, Python!"  
b = 10               
c = 3.14              
d = True              
e = [1, 2, 3]          
f = {"a": 1, "b": 2}    

print(type(a), type(b), type(c), type(d), type(e), type(f))
```

2. Цикли

```python
for i in range(5):
    print(f"Ітерація №{i}")

count = 0
while count < 3:
    print("Лічильник =", count)
    count += 1
```

3. Розгалуження (if / elif / else)

```python
x = 7
if x > 10:
    print("x більше 10")
elif x == 10:
    print("x дорівнює 10")
else:
    print("x менше 10")
```

4. Функції та лямбда-вирази

```python
def greet(name):
    return f"Привіт, {name}!"

print(greet("Анна"))

square = lambda n: n**2
print("Квадрат числа 5:", square(5))
```

5. Обробка помилок

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Помилка:", e)
finally:
    print("Блок finally виконується завжди.")
```

6. Контекст-менеджер

```python
with open("demo.txt", "w", encoding="utf-8") as f:
    f.write("Це тестовий файл для демонстрації контекст-менеджера.\n")

with open("demo.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip())
```
