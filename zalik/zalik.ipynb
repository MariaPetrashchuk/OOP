Створити клас PrimeChecker, який:

- Приймає в конструкторі число n
- Має метод check(), який перевіряє чи є число простим
- Має property is_prime, яка повертає True або False

```python
class PrimeChecker:
    def __init__(self, n):
        """Конструктор приймає число n"""
        self.n = n
        self._is_prime_cached = None
    
    def check(self):
        """Метод перевіряє чи є число простим"""
        if self.n < 2:
            return False
        
        if self.n == 2:
            return True
        
        if self.n % 2 == 0:
            return False
        
        
        for i in range(3, int(self.n ** 0.5) + 1, 2):
            if self.n % i == 0:
                return False
        
        return True
    
    @property
    def is_prime(self):
        """Property, яка повертає True або False"""
        if self._is_prime_cached is None:
            self._is_prime_cached = self.check()
        return self._is_prime_cached
    
    def __str__(self):
        """Метод для виведення інформації про студента"""
        return f"Студентка: Петращук Марія, Група: КН-33"  


print("=== Демонстрація роботи класу PrimeChecker ===\n")

numbers = [17, 20, 2]
for num in numbers:
    checker = PrimeChecker(num)
    print(checker) 
    print(f"Перевірка числа: {num}")
    print(f"Чи є число простим? {checker.is_prime}")
    print(f"Результат методу check(): {checker.check()}")
    print("-" * 40)

```
