Методи класу

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"


a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)

print("\n=== Виклик speak() для всіх тваринок ===")
print(f"{a1.animals}: {a1.speak()}")
print(f"{a2.animals}: {a2.speak()}")
print(f"{a3.animals}: {a3.speak()}")

print("\n=== Додаємо 1 рік кожній тваринці ===")
print(a1.add_one_year())
print(a2.add_one_year())
print(a3.add_one_year())

print("\n=== Новий вік після збільшення ===")
print(f"{a1.animals} тепер {a1.age} років")
print(f"{a2.animals} тепер {a2.age} років")
print(f"{a3.animals} тепер {a3.age} років")
```

Змінні класу, глобальні змінні над всіма обєктами

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass == None else mass
        
        # Оновлюємо глобальні змінні класу
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1


print(f"Спочатку: типи → {list(MyAnimals.total_animal_types)}, кількість → {MyAnimals.total_animals}")

# --- Створюємо об’єкти в циклі ---
animals_data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5),
    ("Собака", "Рекс", 2)
]

objects = []  # список для збереження об’єктів

for a, n, age in animals_data:
    objects.append(MyAnimals(a, n, age))

print("\nПісля створення в циклі:")
print(f"В мене зараз є тваринки → {list(MyAnimals.total_animal_types)}")
print(f"Загальна кількість → {MyAnimals.total_animals}")

# Доступ до змінної класу через об’єкт
a1 = objects[0]
print(f"\nДоступ через об’єкт (a1): {a1.total_animal_types}")

# Присвоєння створює вже НЕ змінну класу, а атрибут об’єкта!
a1.total_animal_types = {"Собака"}
print(f"Після зміни a1.total_animal_types: {a1.total_animal_types}")
print(f"Значення у класі → {MyAnimals.total_animal_types}")
```

Статичні методи та змінна self

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)


print(f"a1 викликає статичний метод: {a1.call_pet()}")
print(f"a2 викликає статичний метод: {a2.call_pet()}")


print(f"Виклик через клас: {MyAnimals.call_pet()}")

print()


print(f"Викликаю метод add_one_year() через об'єкт: {a1.add_one_year()}")
print(f"Новий вік a1: {a1.age}")


print(f"Викликаю add_one_year() через клас: {MyAnimals.add_one_year(a2)}")
print(f"Новий вік a2: {a2.age}")
```

Робота з properties

```python
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
     
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age

   
    @property
    def age_group(self):
        if self.age < 3:
            return "малюк"
        elif self.age < 10:
            return "доросла тваринка"
        else:
            return "стара тваринка"

    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"


a1 = MyAnimals("Кіт", "Мурка", 5)

print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Вікова категорія: {a1.age_group}")

a1.add_one_year()

print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Вікова категорія: {a1.age_group}")
```

Методи Класу

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

    @classmethod
    def is_name_capitalized(cls, name: str) -> bool:
        """Перевіряє, чи починається ім'я з великої літери."""
        return name[0].isupper()


a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

print(MyAnimals.is_name_capitalized(a1.name))  # True
print(MyAnimals.is_name_capitalized("рекс"))   # False
```

Приватні атрибути та методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self._secret_phrase = "Це приватний атрибут, (як я буду кликати свою тваринку)"
        self.__secret_strong = "Цей атрибут точно не можна викликати ззовні!!"
    
    @property
    def secret_phrase(self):
        return f"Ми доступаємось до приватного атрибуту з проперті: {self._secret_phrase}"
    
    @property
    def secret_strong(self):
        return f"Ми доступаємось до приватного атрибуту з проперті: {self.__secret_strong}"

    def _secret_methid(self):
        return "Приватний метод, краще його не рухати, він щось робить потрібне всередині класу"

    def __do_not_touch_it(self):
        return "Це неможливо викликати ззовні"
    


a1 = MyAnimals("Собака", "Рекс", 2)

print(a1._secret_phrase)
a1._secret_phrase = "Я змінив приватний атрибут"  ## Такого робити не треба!
print(a1._secret_phrase)

## a1.__secret_strong ## Це видасть помилку!
print(a1._MyAnimals__secret_strong) ### Так також робити не можна!

### Виводимо приватний атрибут через проперті
print(a1.secret_phrase)
#a1.secret_phrase = "З проперті так зробити не можна! це видасть помилку"

print(a1.secret_strong)

print(a1._secret_methid())

#a1.__do_not_touch_it() ##Це виклик видасть помилку!
```

Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```

```python
a = 2
b = 5
c = "C"

print(a+b, a - int(1),  a * b, str(a) + c)

print(a.__add__(b), a.__sub__(int(1)), a.__mul__(b))

class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"Мій {self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            self.age += obj
            return f"Мій {self.name} постарів на {obj} років, тепер йому {self.age}"
        return NotImplementedError
    
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            if self.age > obj.age:
                return f"Старшим є {self.name}"
            return f"{self.name} не є старшим за {obj.name}"
        return NotImplementedError

    def __len__(self) -> int:
        print("Ми зайшли в метон len, та знаходимо довжину Імені тваринки")
        return len(self.name)
            
        

a = MyAnimals("Собака", "Рекс", 2)
b = MyAnimals("Кіт", "Мурзик", 4)
print(a - b)
print(a - 1)
print(a - float(1))

print(a > b)

print(b > a)

len(a)
```
